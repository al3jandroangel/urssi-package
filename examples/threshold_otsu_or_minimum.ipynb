# An example notebook file showing the threshold_otsu_or_minimum function in action

```python
from urssi_package.filters import threshold_otsu_or_minimum

from skimage import data
import matplotlib.pyplot as plt
```

## Load data and visualize

Load the human mitosis image from scikit-image.

```python
img = data.human_mitosis()
print(img.shape)
plt.imshow(img, cmap='gray')
```

## Get Otsu's threshold valuem

```python
threshold_otsu = threshold_otsu_or_minimum(img)
print(f"threshold value: {threshold_otsu}")
img_otsu = img > threshold_otsu
plt.imshow(img_otsu, cmap='gray')
```

Confirm that setting the minimum value is useful

```python
threshold_otsu_or_minimum(img, min_value=0)
```

```python
threshold_min = threshold_otsu_or_minimum(img, min_value=90)
print(f"threshold value: {threshold_min}")
img_min = img > threshold_min
plt.imshow(img_min, cmap='gray')
```
